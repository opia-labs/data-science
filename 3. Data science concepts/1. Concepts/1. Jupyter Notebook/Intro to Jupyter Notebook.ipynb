### Introduction:
One of the most common question people ask is which IDE / environment / tool to use, while working on our data science projects. As we would expect, there is no dearth of options available – from language specific IDEs like R Studio, PyCharm to editors like Sublime Text or Atom – the choice can be intimidating for a beginner.

If there is one tool which every data scientist should use or must be comfortable with, it is Jupyter Notebooks (previously known as iPython notebooks). Jupyter Notebooks are powerful, versatile, shareable and provide the ability to perform data visualization in the same environment.


#### How to install Jupyter Notebook:
Anaconda:

    Anaconda installs both python and jupyter notebook tools and includes quite a lot of packages commonly used in the data science and machine learning community. 

The pip method:

    If we have Python already installed, pip will already be there. To upgrade to the latest pip version, follow the below code:
    #Linux and OSX : pip install -U pip setuptools
    #Windows : python -m pip install -U pip setuptools
    Once pip is ready, we can go ahead and install Jupyter:
    #For Python2: pip install jupyter
    #For Python3: pip3 install jupyter


#### Getting Started!:
To run Jupyter notebook, simply type “jupyter notebook”. Once we do this, the Jupyter notebook will open up in our default web browser with the below URL:

http://localhost:8888/tree



To open a new Jupyter notebook, click on the ‘New’ option on the right-hand side of the page. Here, we get four options to choose from:

    Python 3
    Text File
    Folder
    Terminal

Text File:It basically works as a text editor
Folder: We can create a new folder to put our documents in, rename it and delete it, whatever our requirement.
Terminal: It works exactly like the terminal on our Mac or Linux machine (cmd on Windows).



#### Open a new Jupyter notebook.
In the drop-down menu we have four options:

Code – This is self-explanatory; it is where we type the code
Markdown – This is where we type the text. We can add conclusions after running a code, add comments, etc.
Raw NBConvert – It’s a command line tool to convert the  notebook into another format (like HTML)
Heading – This is where we add Headings to separate sections and make the notebook look tidy and neat.
